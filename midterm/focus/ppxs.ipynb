```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import t

# Tham số
n = 20  # Số sản phẩm trong 1 lô
p = 0.3  # Xác suất lỗi
size = 10  # Số lô mô phỏng

# 1. rvs: Mô phỏng số lỗi trong 10 lô
errors = binom.rvs(n=n, p=p, size=size)
print("Số sản phẩm lỗi trong 10 lô:", errors)

# 2. pmf: Xác suất có đúng 5 sản phẩm lỗi
prob_5_errors = binom.pmf(k=5, n=n, p=p)
print("Xác suất có đúng 5 sản phẩm lỗi:", prob_5_errors)

# . ppf: Ngưỡng mà 90% số lỗi nằm dưới
threshold_90 = binom.ppf(q=0.9, n=n, p=p)
print("90% số lỗi nhỏ hơn hoặc bằng:", threshold_90)

""" 
Trong trường hợp này:
    q = 0.9: Bạn muốn P(X ≤ k) = 0.9, tức 90% các lô có số lỗi nhỏ hơn hoặc bằng k.
    n = 20, p = 0.3: Đây là tham số của phân phối Nhị Thức.
Kết quả
    Khi chạy đoạn mã, binom.ppf(q=0.9, n=20, p=0.3) sẽ trả về một giá trị k (số lỗi).
    Giả sử kết quả là threshold_90 = 8.
    Ý nghĩa: 90% các lô hàng sẽ có số lỗi nhỏ hơn hoặc bằng 8. """
    
# 4. Vẽ đồ thị PMF
k_values = np.arange(0, n+1)
print(k_values)
pmf_values = binom.pmf(k_values, n=n, p=p)
print(pmf_values)
plt.bar(k_values, pmf_values, color='skyblue')
plt.title("Phân phối Nhị Thức (n=20, p=0.3)")
plt.xlabel("Số sản phẩm lỗi")
plt.ylabel("Xác suất")
plt.show()
```

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

# Tham số
mu = 4  # Tỷ lệ trung bình khách trong 1 giờ
size = 10  # Số giờ mô phỏng

# 1. rvs: Mô phỏng số khách trong 10 giờ
customers = poisson.rvs(mu=mu, size=size)
print("Số khách trong 10 giờ:", customers)

# 2. pmf: Xác suất có đúng 3 khách trong 1 giờ
prob_3 = poisson.pmf(k=3, mu=mu)
print("Xác suất có đúng 3 khách:", prob_3)

# 3. ppf: Ngưỡng mà 95% số khách nằm dưới
threshold_95 = poisson.ppf(q=0.95, mu=mu)
print("95% số khách nhỏ hơn hoặc bằng:", threshold_95)

# 4. Vẽ đồ thị PMF để trực quan hóa
k_values = np.arange(0, 15)  # Phạm vi giá trị để vẽ
pmf_values = poisson.pmf(k_values, mu=mu)
plt.bar(k_values, pmf_values, color='skyblue')
plt.title("Phân phối Poisson (mu=4)")
plt.xlabel("Số khách")
plt.ylabel("Xác suất")
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Tham số
loc = 170  # Trung bình (cm)
scale = 10  # Độ lệch chuẩn (cm)

# 1. rvs: Mô phỏng chiều cao của 1000 người
heights = norm.rvs(loc=loc, scale=scale, size=1000)

# 2. cdf: Tính xác suất chiều cao dưới 160 cm
prob_below_160 = norm.cdf(x=160, loc=loc, scale=scale)
print("Xác suất chiều cao dưới 160 cm:", prob_below_160)

# 3. ppf: Tìm chiều cao mà 99% số người thấp hơn
height_99th = norm.ppf(q=0.99, loc=loc, scale=scale)
print("99% số người thấp hơn:", height_99th, "cm")

# 4. Vẽ đồ thị PDF và histogram
# Đường cong PDF
x = np.linspace(130, 210, 100)  # Phạm vi để vẽ
pdf_values = norm.pdf(x, loc=loc, scale=scale)

# Vẽ histogram của dữ liệu mô phỏng
plt.hist(heights, bins=30, density=True, alpha=0.7, color='skyblue', label='Histogram')
plt.plot(x, pdf_values, 'r-', lw=2, label='PDF')
plt.title("Phân phối Chuẩn: Chiều cao (mean=170, std=10)")
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Tần suất/Mật độ")
plt.legend()
plt.show()
```
