# Bài 2: (3đ)
1. (1đ) Dựa vào số liệu trong quá khứ, ta ước lượng rằng 85% các sản phẩm của một máy sản xuất nào đó là thứ phẩm. Nếu máy này sản xuất 20 sản phẩm mỗi giờ, thì xác suất 48 đến 50 thứ phẩm được sản xuất trong mỗi khoảng thời gian 3 giờ là bao nhiêu?
2. Tỷ lệ một loại bệnh bẩm sinh trong dân số là p = 0.1\%. Bệnh này cần sự chăm
sóc đặc biệt lúc mới sinh. Một trung tâm y tế thường có 500 ca sinh trong một tuần. Tính xác suất để

  (a) (0.5) không có trường hợp nào cần chăm sóc đặc biệt,

  (b) (0.5) có đúng một trường hợp cần chăm sóc đặc biệt,

  (c) (1đ) có nhiều hơn một trường hợp cần chăm sóc đặc biệt.

```python
#1
# Bài 1: Xác suất số thứ phẩm trong 3 giờ
n = 60  # 20 sản phẩm/giờ * 3 giờ
p = 0.85  # Xác suất sản phẩm là thứ phẩm
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

# Xấp xỉ chuẩn với hiệu chỉnh liên tục
z1 = (47.5 - mu) / sigma
z2 = (50.5 - mu) / sigma

p1 = norm.cdf(z1)  # P(X <= 47.5)
p2 = norm.cdf(z2)  # P(X <= 50.5)

prob_48_50 = p2 - p1
print(f"Xác suất có từ 48 đến 50 thứ phẩm: {prob_48_50:.4f}")
```

```python
#2
# Bài 2: Xác suất số ca bệnh cần chăm sóc đặc biệt
n_births = 500  # Số ca sinh mỗi tuần
p_disease = 0.001  # Xác suất mắc bệnh
lambda_poisson = n_births * p_disease  # Trung bình số ca mắc bệnh

# Tính xác suất với phân phối Poisson
p_0 = poisson.pmf(0, lambda_poisson)  # P(X = 0)
p_1 = poisson.pmf(1, lambda_poisson)  # P(X = 1)
p_more_than_1 = 1 - (p_0 + p_1)  # P(X > 1)

print(f"Xác suất không có ca bệnh: {p_0:.4f}")
print(f"Xác suất có đúng 1 ca bệnh: {p_1:.4f}")
print(f"Xác suất có nhiều hơn 1 ca bệnh: {p_more_than_1:.4f}")
```

# Bài 3 (3đ)
Trọng lượng X (tính bằng gam) một loại trái cây có phân phối chuẩn $N(\mu, \sigma^2)$, với $\mu = 500$ (gam) và $\sigma^2 = 16\ (\text{gam}^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

(a) loại 1 : trên 505 gam,

(b) loại 2 : từ 495 đến 505 gam,

(c) loại 3 : dưới 495 gam.

Nhà anh Nam có vườn trái cây thu hoạch được 1000 trái cây. Anh Nam cần chuẩn bị bao đựng trái đây, mỗi bao đựng được khối lượng 25kg. Quy ước làm tròn loại 1 đều cân nặng 550g, loại 2 cân nặng 500g và loại 3 nặng 450g. Hỏi anh Nam cần chuẩn bị bao nhiêu bao đựng cho mỗi loại trái cây?

```python
# Thông số của phân phối chuẩn
mu = 500
sigma = 4
total_fruits = 1000  # Tổng số trái cây

# Tính xác suất
p_loai1 = 1 - norm.cdf(505, mu, sigma)  # X > 505
p_loai2 = norm.cdf(505, mu, sigma) - norm.cdf(495, mu, sigma)  # 495 <= X <= 505
p_loai3 = norm.cdf(495, mu, sigma)  # X < 495

# Tính số trái mỗi loại
n_loai1 = round(total_fruits * p_loai1)
n_loai2 = round(total_fruits * p_loai2)
n_loai3 = round(total_fruits * p_loai3)

# Khối lượng mỗi trái
w_loai1 = 0.55  # kg
w_loai2 = 0.50
w_loai3 = 0.45

# Tính số bao cần thiết (mỗi bao chứa 25 kg)
bags_loai1 = np.ceil(n_loai1 * w_loai1 / 25)
bags_loai2 = np.ceil(n_loai2 * w_loai2 / 25)
bags_loai3 = np.ceil(n_loai3 * w_loai3 / 25)

# In kết quả
print(f"Số lượng trái loại 1 (>505g): {n_loai1}, cần {int(bags_loai1)} bao")
print(f"Số lượng trái loại 2 (495-505g): {n_loai2}, cần {int(bags_loai2)} bao")
print(f"Số lượng trái loại 3 (<495g): {n_loai3}, cần {int(bags_loai3)} bao")
```
