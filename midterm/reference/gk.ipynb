```python
sample = df.sample(n=50, random_state=1)
sample
```

```python
sample_gender_ratio = sample['Gender'].value_counts(normalize=True) 
```

```python
error = abs(gender_ratio['F'] - sample_gender_ratio.get('F', 0)) 
print("Tỷ lệ nữ trong mẫu:", sample_gender_ratio.get('F', 0)) 
print("Sai số so với giá trị ban đầu:", error) 
```

```python
sample_means = [] 
for _ in range(100): 
    sample = df.sample(n=50, random_state=1) 
    sample_gender_ratio = sample['Gender'].value_counts(normalize=True) 
    sample_means.append(sample_gender_ratio['F']) 
    plt.hist(sample_means, bins=20, color='lightblue', edgecolor='black') 
    plt.title('Histogram of Sample Mean Female Ratio') 
    plt.show() 
```

```python
sample_sizes = [30, 50, 100, 200] 
sample_means = [] 
for sample_size in sample_sizes: 
    sample_means_for_size = [] 
    for _ in range(100):  # Lặp lại quá trình lấy mẫu 100 lần 
        sample = df.sample(n=sample_size, random_state=1) 
        sample_gender_ratio = sample['Gender'].value_counts(normalize=True) 
        sample_means_for_size.append(sample_gender_ratio.get('Female', 0)) 
        sample_means.append(sample_means_for_size) 
        plt.figure(figsize=(12, 8)) 
for i, sample_size in enumerate(sample_sizes): 
    plt.subplot(2, 2, i + 1) 
    plt.hist(sample_means[i],bins=20, color='lightblue', edgecolor='black') 
    plt.title(f'Sample Mean Female Ratio (n={sample_size})') 
    plt.tight_layout() 
    plt.show()
```
