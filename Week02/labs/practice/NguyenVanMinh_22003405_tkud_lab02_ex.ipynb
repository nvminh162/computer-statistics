## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

```python
import numpy as np
arr = np.random.randint(0, 1001, size=100)
print(arr)
```

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
print(arr[:10])
print(arr[-5:])
print(arr[(arr % 5 == 0) & (arr > 100)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
print('Mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100', arr[arr < 100].size)
print('Hiển thị các phần tử đó', arr[arr < 100])
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
arr.sort()
print(arr) 

print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))

Q1 = np.percentile(arr, 25)  # Phân vị 25 (Q1)
Q2 = np.median(arr)          # Phân vị 50 (Q2, trung vị)
Q3 = np.percentile(arr, 75)  # Phân vị 75 (Q3)
print(f"Phân vị Q1 (25%): {Q1}")
print(f"Phân vị Q2 (50%): {Q2}")
print(f"Phân vị Q3 (75%): {Q3}")

print(np.percentile(arr, 20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
random_choices = [np.random.choice(arr, 10, replace=False) for _ in range(5)]

# Hiển thị kết quả lựa chọn
for i, choice in enumerate(random_choices, 1):
    print(f"Lần {i}: {choice}")
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
most_common = Counter(arr).most_common(1)[0]

# Hiển thị kết quả
print(f"Phần tử xuất hiện nhiều nhất là {most_common[0]} với {most_common[1]} lần.")
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)', ls='--')
plt.grid(True)
plt.legend()
plt.show()
```
