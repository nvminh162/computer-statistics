1. Viết chương trình nhập vào số nguyên n (n>0). In ra các số nguyên tố từ 1..n


````python
n = int(input("Nhập giá trị của n: "))

for num in range(2, n+1):
  is_prime = True
  for i in range(2, num):
    if num % i == 0:
      is_prime = False
      break
  if is_prime:
    print(num)
````

````python
# Nhập số nguyên n
while True:
    try:
        n = int(input('Nhập vào số nguyên n (n > 0): '))
        if n > 0:
            break
        else:
            print('Vui lòng nhập số nguyên n > 0.')
    except ValueError as e:
        print(f"Vui lòng nhập một số nguyên hợp lệ!\nLỗi: {e}")

# In các số nguyên tố từ 1 đến n
print(f"Các số nguyên tố từ 1 đến {n} là:")
for num in range(2, n + 1):  # Bắt đầu từ 2 vì 1 không phải số nguyên tố
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):  # Kiểm tra từ 2 đến √num
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
````

2. Viết chương trình nhập vào số nguyên n có 3 chữ số. Tính tổng các chữ số của n

````python
def totalDigitsOfNumber(n):
    total = 0
    while (n > 0):
        total = total + n % 10;
        n = int(n // 10);
    return total;
 
n = int(input("Nhập số nguyên dương n = "));
print("Tổng các chữ số của", n , "là", totalDigitsOfNumber(n));
````

3. Viết chương trình nhập vào một dãy các số nguyên được cách nhau bởi khoảng trắng (VD: 32 8 4 6 9...) lưu vào python ở dạng mảng. Thực hiện

- Xuất mảng
- Tính tổng các phần tử của mảng
- Tìm số lớn nhất
- Tìm số bé nhất
- Sắp xếp mảng theo thứ tự giảm dần
- In ra các số chẵn trong mảng

````python
# Nhập vào một dãy các số nguyên cách nhau bởi khoảng trắng
numbers = list(map(int, input("Nhập dãy số nguyên: ").split()))

# Xuất mảng
print("Mảng đã nhập:", numbers)

# Tính tổng các phần tử của mảng
total = sum(numbers)
print("Tổng các phần tử trong mảng:", total)

# Tìm số lớn nhất
max_num = max(numbers)
print("Số lớn nhất trong mảng:", max_num)

# Tìm số bé nhất
min_num = min(numbers)
print("Số bé nhất trong mảng:", min_num)

# Sắp xếp mảng theo thứ tự giảm dần
numbers.sort(reverse=True)
print("Mảng sau khi sắp xếp theo thứ tự giảm dần:", numbers)

# In ra các số chẵn trong mảng
even_numbers = [num for num in numbers if num % 2 == 0]
print("Các số chẵn trong mảng:", even_numbers)
````

4. Viết chương trình yêu cầu người dùng nhập họ tên. Hiển thị tên người dùng sau khi đã chuẩn hóa. Biết rằng tên đã chuẩn hóa là:
- Ký tự đầu viết hoa
- Không có khoảng trắng thừa
- Không có ký tự đặc biệt

````python
import re

# Hàm chuẩn hóa tên
def normalize_name(name):
    # Loại bỏ khoảng trắng thừa ở đầu và cuối
    name = name.strip()
    # Thay thế các khoảng trắng liên tiếp bằng một khoảng trắng duy nhất
    name = re.sub(r'\s+', ' ', name)
    # Viết hoa ký tự đầu của mỗi từ và viết thường các ký tự còn lại
    name = name.title()
    # Loại bỏ các ký tự đặc biệt
    name = re.sub(r'[^a-zA-Z\s]', '', name)
    return name

# Nhập họ tên từ người dùng
full_name = input("Nhập họ tên: ")

# Chuẩn hóa tên và hiển thị
normalized_name = normalize_name(full_name)
print("Tên đã chuẩn hóa:", normalized_name)
````

5. Viết chương trình nhập vào một số nguyên dương (n>0). Cho biết n có phải là số đối xứng hay không? Biết số đối xứng là số viết xuôi hay viết ngược đều như nhau

````python
# Hàm kiểm tra số đối xứng
def is_palindrome(n):
    # Chuyển số thành chuỗi và so sánh với chuỗi đảo ngược
    return str(n) == str(n)[::-1]

# Nhập số nguyên dương từ người dùng
n = int(input("Nhập một số nguyên dương (n > 0): "))

# Kiểm tra và thông báo kết quả
if n > 0:
    if is_palindrome(n):
        print(f"{n} là số đối xứng.")
    else:
        print(f"{n} không phải là số đối xứng.")
else:
    print("Vui lòng nhập một số nguyên dương n > 0.")
````

````python
# Tìm các số chia hết cho 7 nhưng không phải bội số của 5 trong đoạn từ 2000 đến 3200
result = [str(i) for i in range(2000, 3201) if i % 7 == 0 and i % 5 != 0]

# In các số tìm được, cách nhau bằng dấu phẩy
print(",".join(result))
````

9. Viết hàm nhận vào một danh sách các số nguyên. Hãy viết hàm trả về một danh sách gồm các số không trùng nhau trong danh sách ban đầu

````python
def remove_duplicates(nums):
    return list(set(nums))

# Ví dụ sử dụng hàm
input_list = [1, 2, 2, 3, 4, 4, 5, 5, 6]
output_list = remove_duplicates(input_list)

print("Danh sách ban đầu:", input_list)
print("Danh sách không trùng nhau:", output_list)
````

10. Viết hàm có tham số đầu vào là một list L có các phần tử là chuỗi. Hãy tạo ra một dictionary D mã hóa, với mỗi một phần tử trong L được mã hóa thành một con số (theo thứ tự từ 0 tăng dần lên 1 đơn vị). Sau đó trả về list đã được mã hóa

    Ví dụ:

    ```
    Cho L = ["đen","vàng","xanh","vàng","xanh","đỏ","hồng"]

    Xây dựng dictionary mã hóa:
    D = {'đen':0,'vàng':1,'xanh':2,'đỏ':3,'hồng':4}

    Trả về List mã hóa:
    L_mahoa = [0, 1, 2, 1, 2, 3, 4]
    ```

````python
def ma_hoa_list(L):
    D = {value: index for index, value in enumerate(sorted(set(L)))}
    L_mahoa = [D[element] for element in L]
    return L_mahoa

L = ["đen", "vàng", "xanh", "vàng", "xanh", "đỏ", "hồng"]
L_mahoa = ma_hoa_list(L)
print(L_mahoa)
````
